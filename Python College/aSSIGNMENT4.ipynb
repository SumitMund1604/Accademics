10. Find the sum of the series 2 +22 + 222 + 2222 + .. n terms.  in python  with list we have to display like that

```python
n = int(input("Enter number of terms: "))

terms = []
display_parts = []

for i in range(1, n + 1):
    num_str = '2' * i          
    num = int(num_str)
    terms.append(num)
    display_parts.append(num_str)

series_display = " + ".join(display_parts)

total = sum(terms)

print("\nSeries:")
print(series_display)

print("\nSum =", total)
```

11. Given a two list. Create a third list by picking an odd-index element from the first list and even

Given a list slice it into a 3 equal chunks and reverse each list.

```python
lst = [1, 2, 3, 4, 5, 6, 7, 8, 9]  
if len(lst) % 3 != 0:
    print("List length must be divisible by 3")
else:
    chunk_size = len(lst) // 3
    chunks = [lst[i:i+chunk_size] for i in range(0, len(lst), chunk_size)]
    reversed_chunks = [chunk[::-1] for chunk in chunks]
    print(reversed_chunks)
```

Given a two list of equal size create a set such that it shows the element from both lists in the

19. Count all lower case, upper case, digits, and special symbols from a given string. 


```python
s1 = input("Enter s1")
u=0
l=0
dig=0
special=0
for char in s1:
    if char.islower():
        l+=1
    elif char.isupper():
        u+=1
    elif char.isdigit():
        dig+=1
    else:
        special+=1
        
print("lower: ",l)
print("uppwe: ",u)
print("digit: ",dig)
print("special: ",special)
```

. Given a string, return the sum and average of the digits that appear in the string, ignoring all 
other characters.

```python
s1 = input("Enter s1")
sum=0
count=0
for char in s1:
    if char.isdigit():
          sum+=int(char)
          count+=1
    
print("avrage",sum/count)
```
