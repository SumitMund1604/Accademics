4. Find the number of occurrences for any repeated item present in a tuple. 


```python
from collections import Counter
my_tuple = (1, 2, 2, 3, 4, 2, 1, 4, 5, 4)
item_counts = Counter(my_tuple)

print("Repeated items and their occurrences:")
for item, count in item_counts.items():
    if count > 1:
        print(f"Item {item} appears {count} times")
```

 16. Sort a tuple by its float element. Sample data: [('item1', '12.20'), ('item2',
 '15.10'), ('item3', '24.5')] Output: [('item3', '24.5'), ('item2', '15.10'),
 ('item1', '12.20')] 

```python
my_tuple = [('item1', '12.20'), ('item2', '15.10'), ('item3', '24.5')]

def get_price(item):
    return float(item[1])

sorted_tuple = sorted(my_tuple, key=get_price)
print(sorted_tuple)  
```

17. Count the number of elements in a tuple until an element is a list. 


```python
my_tuple = tuple(eval(input("Enter the tuple separated by comma: ")))
count=0
for t in my_tuple:
    if type(t)==list:
        break
    count+=1

print(count)
```
