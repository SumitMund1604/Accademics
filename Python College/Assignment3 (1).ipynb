1.	Write a program to find the smallest and largest digit in a number.

```python
number = int(input("Enter a  Number:"))


temp_num = number
last_digit = temp_num % 10
smallest = last_digit
largest = last_digit

while temp_num > 0:
    digit = temp_num % 10  

    if digit < smallest:
        smallest = digit

    if digit > largest:
        largest = digit

    temp_num //= 10  

print(f"The given number is: {number}")
print(f"The smallest digit is: {smallest}")
print(f"The largest digit is: {largest}")
```

13.	 Swap the first and last digit of a number.

```python
number = int(input("enter a no: "))
last_digit = number % 10
power_of_10 = 1
temp_num = number
while temp_num >= 10:
    temp_num //= 10
    power_of_10 *= 10

first_digit = number // power_of_10

md = number % power_of_10 // 10

swapped_number = (last_digit * power_of_10) + (md * 10) + first_digit

print(f"Swapped number is {swapped_number}")
```

15.	Check whether a number is a Harshad number (divisible by sum of digits).

```python
num = int(input("Enter a no: ")) 

temp = num
sum_no = 0

while temp > 0:
    digit = temp % 10
    sum_no += digit
    temp //= 10

if sum_no != 0 and num % sum_no == 0:
    print(f"{num} is divisible by its sum of digits ({sum_no})")
else:
    print(f"{num} is not divisible by its sum of digits ({sum_no})")
```

17.	Find numbers between 1 and 1000 that become prime after removing one digit.

```python
for n in range(10, 1000):
    is_valid = False
    
    temp_n = n
    digit_count = 0
    while temp_n > 0:
        temp_n //= 10
        digit_count += 1
        
    for i in range(digit_count):
        power = 10**i
        left = n // (power * 10)
        right = n % power
        val = (left * power) + right
        
        if val > 1:
            is_prime = True
            d = 2
            while d * d <= val:
                if val % d == 0:
                    is_prime = False
                    break
                d += 1
            
            if is_prime:
                is_valid = True
                break 
                
    if is_valid:
        print(n, end=", ")
```
